```python
import glob
import numpy as np
from pprint import pprint 
folder=glob.glob("./lol/*")
folder=sorted(folder, key=lambda s: s.lower())
for f in folder:
    print f
print len(folder)
```

```python
price=np.loadtxt('price.txt')
print price
```

```python
from PIL import Image
im=[]
p=[]
w=[]
h=[]
c=[]
q=0
for f in folder:
    files=glob.glob(f+'/*.jpg')
    for j in files:
        img=Image.open(j)
        #print img.size
        img=img.resize((30,30), Image.ANTIALIAS)
        newim=np.asarray(img)
        '''break
        b=plt.imread(i)
        '''
        b=newim.flatten()
        im.append(b)
        p.append(price[q][0])
        w.append(price[q][1])
        h.append(price[q][2])
        c.append(price[q][3])
    q=q+1

print im[0],p[0],w[0],h[0],c[0]
```

```python
print len(im)
from sklearn.svm import SVR
svr = SVR(kernel='rbf', C=1e3, gamma=0.1)
```

```python
oe1=0.0
oe2=0.0
size=len(im)
part=size/4
for i in range(4):
    from keras.models import Sequential
    from keras.layers import Dense, Activation
    m=Sequential()
    m.add(Dense(600,input_dim=2700))
    m.add(Activation('relu'))
    m.add(Dense(100))
    m.add(Activation('relu'))
    m.add(Dense(1))
    m.add(Activation('relu'))

    m.compile(loss='mean_absolute_percentage_error',optimizer='sgd')
    
    #print len(im)
    from sklearn.svm import SVR
    svr = SVR(kernel='rbf', C=1e3, gamma=0.1)

    
    trainx=[]
    trainy=[]
    testx=[]
    testy=[]
    for j in range(size):
        if j < i*part or j>=i*part + part:
            temp=[]
            for l in range(2700):
                temp.append(float(im[j][l])/255)
            trainx.append(temp)
            trainy.append([p[j]])
    for j in range(i*part,i*part + part):
        temp=[]
        for l in range(2700):
            temp.append(float(im[j][l])/255)
        testx.append(temp)
    #testx=[im[x]/255 for x in range(i*175,i*175 + 175)]
    for j in range(i*part,(i+1)*part):
        testy.append(p[j])
    #testy=[p[x] for x in range(i*175,i*175 + 175)]    
    print np.shape(trainx)
    m.fit(trainx,trainy, epochs=10, batch_size=32)
    tp=m.predict(trainx,batch_size=16)
    #print len(tp)
    newtrainx=[]
    k=0
    #print len(tp)
    for j in range(size):
        if j < i*part or j>=(i+1)*part:
            b=[tp[k][0]]
            b.append(w[j])
            b.append(h[j])
            b.append(c[j])
            newtrainx.append(b)
            k=k+1
    svr.fit(newtrainx,trainy)
    tp=m.predict(testx,batch_size=32)
    acc=0.0
    #print ty
    for j in range(len(tp)):
        acc+=(abs(testy[j]-tp[j])/testy[j])*100
    acc=acc/len(tp)
    print acc
    oe1+=acc
    
    newtest=[]
    k=0
    for j in range(i*part,(i+1)*part):
        b=[tp[k][0]]
        k=k+1
        b.append(w[j])
        b.append(h[j])
        b.append(c[j])
        newtest.append(b)
    ty=svr.predict(newtest)
    acc=0.0
    #print ty
    for j in range(len(ty)):
        acc+=(abs(testy[j]-ty[j])/testy[j])*100
    acc=acc/len(ty)
    print acc
    oe2+=acc
```
